```python
import pandas as pd
import matplotlib.pyplot as plt

# Ruta del archivo
file_path = r"D:/Power BI + AI/Informe_Ventas.xlsx"

# Cargar el archivo Excel
df = pd.read_excel(file_path, engine='openpyxl')

# Calcular el monto total de ventas
df['MontoVenta'] = df['Quantity'] * df['UnitPrice']

# Agrupar por país y ordenar de forma descendente
ventas_por_pais = df.groupby('Country')['MontoVenta'].sum().sort_values(ascending=False)

# Seleccionar los 5 países con mayores ventas
top_5_paises = ventas_por_pais.head(5)

# Crear gráfico de torta
plt.figure(figsize=(8, 6))
plt.pie(top_5_paises, labels=top_5_paises.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 países por monto total de ventas')
plt.axis('equal')  # Para que el gráfico sea circular
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV (ajusta el nombre si es diferente)
df = pd.read_csv("ventas.csv", parse_dates=['InvoiceDate'])

# Estadísticas descriptivas
desc_stats = df.describe(include='all')
print("Estadísticas descriptivas:")
print(desc_stats)

# Función para detectar outliers usando IQR
def detectar_outliers(df, columna):
    q1 = df[columna].quantile(0.25)
    q3 = df[columna].quantile(0.75)
    iqr = q3 - q1
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr
    outliers = df[(df[columna] < limite_inferior) | (df[columna] > limite_superior)]
    return outliers

# Detectar outliers
outliers_quantity = detectar_outliers(df, 'Quantity')
outliers_unitprice = detectar_outliers(df, 'UnitPrice')

# Mostrar resultados
print("\nOutliers en Quantity:")
print(outliers_quantity)

print("\nOutliers en UnitPrice:")
print(outliers_unitprice)

# Visualización con boxplots
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x=df['Quantity'])
plt.title('Boxplot de Quantity')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['UnitPrice'])
plt.title('Boxplot de UnitPrice')

plt.tight_layout()
plt.show()
``
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV (ajusta el nombre si es diferente)
df = pd.read_excel("D:/Power BI + AI/Informe_Ventas.xlsx", engine='openpyxl')

# Estadísticas descriptivas
desc_stats = df.describe(include='all')
print("Estadísticas descriptivas:")
print(desc_stats)

# Función para detectar outliers usando IQR
def detectar_outliers(df, columna):
    q1 = df[columna].quantile(0.25)
    q3 = df[columna].quantile(0.75)
    iqr = q3 - q1
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr
    outliers = df[(df[columna] < limite_inferior) | (df[columna] > limite_superior)]
    return outliers

# Detectar outliers
outliers_quantity = detectar_outliers(df, 'Quantity')
outliers_unitprice = detectar_outliers(df, 'UnitPrice')

# Mostrar resultados
print("\nOutliers en Quantity:")
print(outliers_quantity)

print("\nOutliers en UnitPrice:")
print(outliers_unitprice)

# Visualización con boxplots
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x=df['Quantity'])
plt.title('Boxplot de Quantity')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['UnitPrice'])
plt.title('Boxplot de UnitPrice')

plt.tight_layout()
plt.show()
```

```python

# Filtrar datos inválidos
df = df[(df['UnitPrice'] > 0) & (df['Quantity'] > 0)]

# Crear columna de mes
df['Mes'] = df['InvoiceDate'].dt.to_period('M')

# 1️⃣ Evolución de ventas por mes
ventas_mensuales = df.groupby('Mes').apply(lambda x: (x['Quantity'] * x['UnitPrice']).sum())
ventas_mensuales.index = ventas_mensuales.index.to_timestamp()

# 2️⃣ Top 10 productos más vendidos
top_productos = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)

# 3️⃣ Distribución de precios unitarios
precios = df['UnitPrice']

# 📊 Visualización
plt.figure(figsize=(18, 5))

# Gráfico 1: Evolución de ventas por mes
plt.subplot(1, 3, 1)
ventas_mensuales.plot(kind='line', marker='o')
plt.title('Evolución de Ventas por Mes')
plt.xlabel('Mes')
plt.ylabel('Ventas Totales')
plt.xticks(rotation=45)

# Gráfico 2: Top 10 productos más vendidos
plt.subplot(1, 3, 2)
top_productos.plot(kind='barh', color='skyblue')
plt.title('Top 10 Productos Más Vendidos')
plt.xlabel('Cantidad Vendida')
plt.gca().invert_yaxis()

# Gráfico 3: Distribución de precios unitarios
plt.subplot(1, 3, 3)
sns.histplot(precios, bins=50, kde=True, color='salmon')
plt.title('Distribución de Precios Unitarios')
plt.xlabel('Precio Unitario')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```
